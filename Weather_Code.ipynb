```python
#Fetching Weather Data for the JFK Location for years 2019,2020--
#we will repeat this process for the remaining years and locations by modifying the values accordingly
import requests
import pandas as pd
import time
import calendar
#28.426901098825073, -81.31249734360168--mco
#40.6446,-73.7797 - jfk
#41.98043450764176, -87.91257800108377 -- chicago ORD
#43.1113, -76.1143 -- Syracuse SY

# Define the base URL and parameters
base_url = "https://api.weatherbit.io/v2.0/history/hourly"
params = {
    "lat": 40.6446,
    "lon": -73.7797,
    "tz": "local",
    "key": "82559218ca6d48d9892cd1b829d10263"
}

field_list = ['datetime', 'timestamp_local', 'temp', 'app_temp', 'clouds', 'precip', 'rh', 'wind_spd', 'wind_gust_spd', 'wind_dir', 'weather', 'pres', 'slp', 'vis', 'snow']

# Initialize an empty list to store dataframes
dfs = []

# Iterate over the years 2019 and 2020
for year in range(2019, 2021):
    # Iterate over each month
    for month in range(1, 13):
        # Determine the number of days in the month
        num_days = calendar.monthrange(year, month)[1] + 1
        
        # Start and end dates for the month
        start_date = f"{year}-{month:02d}-01"
        end_date = f"{year}-{month:02d}-{num_days:02d}"
        params["start_date"] = start_date
        params["end_date"] = end_date

        # Make the API request
        response = requests.get(base_url, params=params)

        # Check if the request was successful
        if response.status_code == 200:
            # Convert the JSON response to a DataFrame
            json_data = response.json()

            # Extract specific fields from the JSON data
            data_subset = [{key: entry['weather']['description'] if key == 'weather' else entry.get(key) for key in field_list} for entry in json_data['data']]

            # Create DataFrame
            df = pd.DataFrame(data_subset)
            

            # Append DataFrame to list
            dfs.append(df)
        else:
            print(f"Error for {start_date} - {end_date}: {response.status_code}")

        # Introduce a time gap after each month to respect the rate limit
        time.sleep(2)  # Introduce a 2-second delay

# Concatenate all DataFrames into a single DataFrame
weather_df = pd.concat(dfs, ignore_index=True)

# Save the DataFrame to a CSV file
weather_df.to_csv('jfk_weather_data_2021_2022.csv', index=False)

print("Data saved to 'jfk_weather_data_2021_2022.csv")
```

## Obtaining the forecast data for the period from April 19th to April 23rd as well.

```python
#Obtaining the forecast data for the period from April 19th to April 23rd as well.
#28.426901098825073, -81.31249734360168--mco
#40.6446,-73.7797 - jfk
#41.98043450764176, -87.91257800108377 -- chicago ORD

import requests
import pandas as pd
import time

# Define the base URL for forecast data and parameters
forecast_base_url = "https://api.weatherbit.io/v2.0/forecast/hourly"
forecast_params = {
    "lat": 28.426901098825073,
    "lon": -81.31249734360168,
    "key": "b843dfc4409d4598a42ec00ffd2fc5ec",
    "start_date": "2024-04-19",
    "end_date": "2024-04-23"
}

field_list = ['datetime', 'timestamp_local', 'temp', 'app_temp', 'clouds', 'precip', 'rh', 'wind_spd', 'wind_gust_spd', 'wind_dir', 'weather', 'pres', 'slp', 'vis', 'snow']

# Make the forecast data API request
forecast_response = requests.get(forecast_base_url, params=forecast_params)

# Check if the request was successful
if forecast_response.status_code == 200:
    # Convert the JSON response to a DataFrame
    forecast_json_data = forecast_response.json()

    # Extract specific fields from the JSON data
    forecast_data_subset = [{key: entry['weather']['description'] if key == 'weather' else entry.get(key) for key in field_list} for entry in forecast_json_data['data']]

    # Create DataFrame
    forecast_weather_df = pd.DataFrame(forecast_data_subset)

    # Save the DataFrame to a CSV file
    forecast_weather_df.to_csv('mco_forecast_weather_data_2024_04_19_to_22.csv', index=False)

    print("Forecast data saved to 'mco_forecast_weather_data_2024_04_19_to_22.csv'")
else:
    print(f"Error: {forecast_response.status_code}")
```

### Consolidating all years of data by organizing them within a designated folder.

```python
import pandas as pd
import os

# Specify the directory containing the CSV files
directory_path = 'C:/Users/gsaiv/Documents/IML_Project/Weather_data/forecast'

# Create a list to hold all the individual DataFrames
all_dataframes = []

# Loop through each file in the directory
for filename in os.listdir(directory_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory_path, filename)
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        # Append the modified DataFrame to the list
        all_dataframes.append(df)

# Concatenate all modified DataFrames in the list
merged_df = pd.concat(all_dataframes, ignore_index=True)

# Specify the path for the merged CSV file
merged_file_path = 'C:/Users/gsaiv/Documents/IML_Project/Weather_data/forecast/merged_forecast_data.csv'

# Write the merged DataFrame to a new CSV file
merged_df.to_csv(merged_file_path, index=False)

print(f'Merged data saved to {merged_file_path}')
```
